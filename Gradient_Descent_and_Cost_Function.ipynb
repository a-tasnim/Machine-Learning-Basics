```python
import numpy as np
```

```python
def grad_descent(x,y):
  m_curr = b_curr = 0
  iter = 10000
  n = len(x)
  learning_rate = 0.08
  for i in range(iter):
    y_pred = m_curr * x + b_curr
    cost = (1/n) * sum([j**2 for j in (y-y_pred)])
    md = -(2/n) * sum(x*(y-y_pred))
    bd = -(2/n) * sum(y-y_pred)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd
    print(f"m->{m_curr}, b->{b_curr},cost{cost} iterration->{i}")
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
grad_descent(x,y)
```

```python
import math
def gradient_descent(x, y):
  m_curr = b_curr = 0
  learning_rate = 0.0002
  n = len(x)
  itr = 100000
  cost_previous = 0


  for i in range(itr):
    y_pred = m_curr * x + b_curr
    cost = (1/n) * sum(y-y_pred)

    dm = -(2/n) * sum(x*(y-y_pred))
    db = -(2/n) * sum(y-y_pred)
    m_curr = m_curr - learning_rate * dm
    b_curr = b_curr - learning_rate * db

    if math.isclose(cost, cost_previous, rel_tol=1e-20):
      break
    cost_previous = cost
    print(f"m->{m_curr}, b->{b_curr},cost{cost} iterration->{i}")
  return m_curr,b_curr


```

```python
coef,intercept = gradient_descent(np.array(df.math), np.array(df.cs))
print(f"Using gradient descent--> coef: {coef}, intercept: {intercept}")
```
